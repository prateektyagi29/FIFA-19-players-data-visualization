### Replacing missing values in the required columns

```python
fifa.isnull().sum()  # Finding the missing values in the dataset.
```

```python
fifa['International Reputation'].isnull().sum()
```

```python
fifa['International Reputation'].fillna(fifa['International Reputation'].mean(), inplace = True)

# Replacing the missing values with their mean value.
# Mean value substitution is mainly used when the column is numerical.
```

```python
fifa['International Reputation'].isnull().sum()  # Checking if there is still any missing value after substitution.
```

```python
fifa['Release Clause'].isnull().sum()  # Checking if there is any missing value. 
```

```python
for column in ['Release Clause']:
    fifa[column].fillna(fifa['Release Clause'].mode()[0], inplace=True)
    
# Replacing the missing values with the mode.
# Mode value substitution is mainly used when the column is categorical
```

```python
fifa['Release Clause'].isnull().sum()  # Checking if there is still any missing value after substitution. 
```

```python
fifa['Club'].isnull().sum()  # Checking if there is any missing value. 
```

```python
fifa['Club'].fillna('No Club', inplace = True) 

#Replacing the missing values with the variable "No CLub".
```

```python
fifa['Club'].isnull().sum()  # Checking if there is still any missing value after substitution.
```

###  Correlation Heatmap

```python
plt.rcParams['figure.figsize']=(25,15)
hm=sns.heatmap(fifa[['Age', 'Overall', 'Potential', 'Special',
    'Body Type', 'Position',
    'Height', 'Weight', 'Crossing',
    'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
    'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
    'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
    'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
    'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
    'Marking', 'StandingTackle', 'SlidingTackle']].corr(), annot = True, linewidths=.5, cmap='Blues')

hm.set_title(label='Heatmap of dataset', fontsize=20)


# Correlation Heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors
# Or we can also say that these Heat maps display numeric tabular data where the cells are colored depending upon the contained value. 
# Heat maps are great for making trends in this kind of data more readily apparent, particularly when the data is ordered and there is clustering.
```

#### Height vs Acceleration

```python
plt.xlabel('Height', fontsize = 20)
plt.ylabel('Acceleration', fontsize = 20)
plt.title('Height vs Acceleration', fontsize = 25)
sns.barplot(x='Height',y='Acceleration',data=fifa.sort_values('Height', inplace=False), alpha=0.6)
```

```python
# From this barplot we can clearly analyze that, the players with or around the height of 5'2 has better acceleration ability than the other players.
# While the taller players has lesser acceleration than the shorter ones.
# We can deduce that the Height is inversely proportional to the acceleration skill of the player.
```

#### International Reputation of the players

```python
# plotting a pie chart to represent share of international reputation
# A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. 
# In a pie chart, the arc length of each slice is proportional to the quantity it represents. 
# A pie chart is the most simple and effective way to visualize a data.

labels = ['1','2','3','4','5']
print(type(labels))
print(labels)
sizes = list(fifa['International Reputation'].value_counts())[0:5]
colors = ['black', 'brown', 'yellow', 'red', 'darkblue']
explode = [0.1, 0.1, 0.2, 0.5, 0.9]

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie (sizes, labels = labels, colors = colors, explode = explode, shadow = True)
plt.title ('International Reputation of Football Players', fontsize = 20)
plt.legend()
plt.legend(bbox_to_anchor=(0.01,1.1))
plt.show()
```

```python
# From this pie chart we can clearly analyze that the maximum players have reputation of 1, while the players with reputation of 5 are very less.
```

```python
# Considering all the relevant skills of these two for the effective and unbiased comparison.

skills = ['Overall', 'Potential', 'Crossing',
   'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
   'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
   'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
   'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
   'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
   'Marking', 'StandingTackle', 'SlidingTackle']
```

```python
kevin = fifa.loc[fifa['Name'] == 'K. De Bruyne']
kevin = pd.DataFrame(kevin, columns = skills)
modric = fifa.loc[fifa['Name'] == 'L. Modrić']
modric = pd.DataFrame(modric, columns = skills)
```

```python
# Constructing a point plot for their comparison.
# Point plot can be used for clear and simple visualization of a small to medium data.

plt.figure(figsize = (20,10))
sns.pointplot(data=kevin, color = 'blue', alpha=0.6)
sns.pointplot(data=modric, color = 'red', alpha=0.6)
plt.text(5,55,'kevin', color = 'blue', fontsize = 25)
plt.text(5,50, 'modric', color = 'red', fontsize = 25)
plt.xticks(rotation=90)
plt.xlabel('Skills', fontsize = 25)
plt.ylabel('Skill Value', fontsize = 25)
plt.title('Kevin vs Modric', fontsize = 30)
plt.grid()
```

```python
# This plot clearly shows which player of the two has an edge over the other in which department.
# You can easily look at the graph an analyze it. 
```

#### Top 10 clubs with overall best players

```python
# Now getting a list of clubs which owns the best rated player (based on the overall rating of the player).

top_clubs = fifa.groupby(['Club'])['Overall'].max().sort_values(ascending = False)
top_clubs.head(10)
```

```python
# This does not means that these clubs are the best in fifa.
# Eg., the rating of Cristiano Ronaldo is 94 in Fifa 19 and he plays for Juventus, therefore, Juventus is at first place, solely for owning the best player in the world.
```

#### Top Clubs of FIFA 19
